# Mô tả thuộc tính của tập dữ liệu student-mat.csv và student-por.csv

## Thuộc tính chung

- **school**: Trường học của học sinh  
  → Nhị phân: `GP` - Gabriel Pereira hoặc `MS` - Mousinho da Silveira

- **sex**: Giới tính của học sinh  
  → Nhị phân: `F` - nữ hoặc `M` - nam

- **age**: Tuổi của học sinh (số)  
  → Từ 15 đến 22

- **address**: Loại địa chỉ nhà của học sinh  
  → Nhị phân: `U` - thành thị hoặc `R` - nông thôn

- **famsize**: Quy mô gia đình  
  → Nhị phân: `LE3` - ít hơn hoặc bằng 3 người, `GT3` - lớn hơn 3 người

- **Pstatus**: Tình trạng sống chung của cha mẹ  
  → Nhị phân: `T` - sống cùng nhau hoặc `A` - sống ly thân

- **Medu**: Trình độ học vấn của mẹ (số)  
  → 0 - không học, 1 - tiểu học (lớp 4), 2 – lớp 5 đến 9, 3 – trung học phổ thông, 4 – đại học/cao đẳng

- **Fedu**: Trình độ học vấn của cha (số)  
  → Giống như `Medu`

- **Mjob**: Nghề nghiệp của mẹ  
  → `teacher`, `health`, `services`, `at_home`, `other`

- **Fjob**: Nghề nghiệp của cha  
  → Giống như `Mjob`

- **reason**: Lý do chọn trường  
  → `home`, `reputation`, `course`, `other`

- **guardian**: Người giám hộ  
  → `mother`, `father`, `other`

- **traveltime**: Thời gian đi học (số)  
  → 1 - <15 phút, 2 - 15 đến 30 phút, 3 - 30 phút đến 1 giờ, 4 - >1 giờ

- **studytime**: Thời gian học hàng tuần (số)  
  → 1 - <2 giờ, 2 - 2 đến 5 giờ, 3 - 5 đến 10 giờ, 4 - >10 giờ

- **failures**: Số lần trượt môn trước đây (số)  
  → n nếu 1≤n<3, ngược lại là 4

- **schoolsup**: Hỗ trợ học thêm tại trường  
  → `yes` hoặc `no`

- **famsup**: Hỗ trợ học thêm từ gia đình  
  → `yes` hoặc `no`

- **paid**: Học thêm có trả phí trong môn học  
  → `yes` hoặc `no`

- **activities**: Tham gia hoạt động ngoại khóa  
  → `yes` hoặc `no`

- **nursery**: Có từng học mẫu giáo hay không  
  → `yes` hoặc `no`

- **higher**: Có mong muốn học đại học  
  → `yes` hoặc `no`

- **internet**: Có truy cập Internet ở nhà  
  → `yes` hoặc `no`

- **romantic**: Đang trong mối quan hệ tình cảm  
  → `yes` hoặc `no`

- **famrel**: Mối quan hệ trong gia đình (số)  
  → Từ 1 (rất tệ) đến 5 (rất tốt)

- **freetime**: Thời gian rảnh sau giờ học (số)  
  → Từ 1 (rất ít) đến 5 (rất nhiều)

- **goout**: Tần suất đi chơi với bạn bè (số)  
  → Từ 1 (rất ít) đến 5 (rất nhiều)

- **Dalc**: Mức độ uống rượu vào ngày trong tuần (số)  
  → Từ 1 (rất ít) đến 5 (rất nhiều)

- **Walc**: Mức độ uống rượu vào cuối tuần (số)  
  → Từ 1 (rất ít) đến 5 (rất nhiều)

- **health**: Tình trạng sức khỏe hiện tại (số)  
  → Từ 1 (rất tệ) đến 5 (rất tốt)

- **absences**: Số ngày nghỉ học (số)  
  → Từ 0 đến 93

---

## Điểm số liên quan đến môn học

- **G1**: Điểm kỳ I (số)  
  → Từ 0 đến 20

- **G2**: Điểm kỳ II (số)  
  → Từ 0 đến 20

- **G3**: **Điểm cuối kỳ - Mục tiêu cần dự đoán**

```python
from sklearn.model_selection import train_test_split , cross_val_score , GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error , r2_score 
```

```python
df_final = df.copy()
X = df_final.drop('G3' , axis = 1)
y = df_final['G3']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= 0.2 , random_state= 2)
```

```python
models = {
    'Linear' : LinearRegression(),
    'RandomForest' : RandomForestClassifier() , 
    'SVM' : SVR() 
}
result = []
for name , model in models.items():
    score = np.mean(cross_val_score(model , X , y , cv = 5 , scoring= 'r2' ))
    result.append({'model' : name , 'score' : score})
df_r2 = pd.DataFrame([{'name' : r['model'] , 'score' : r['score']} for r in result])
df_r2 = df_r2.sort_values(by = 'score' , ascending= False , ignore_index= True)
df_r2
```

```python
sns.pointplot(x = df_r2['name'] , y = df_r2['score'])
plt.ylim(0.5 , 0.8)
```

# Cải tiến mô hình 

```python
from math import * 
result = []
models = {
    'Linear' : LinearRegression(),
    'SVM' : SVR(kernel='linear') 
}
param_grid= {
    'SVM' : {
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'epsilon': [0.01, 0.1, 0.2],
        'max_iter': [1000, 5000, -1]
}
}
for name , model in models.items() : 
    if name in param_grid : 
        grid_search = GridSearchCV(model , param_grid[name] , cv = 5 , scoring = 'r2')
        grid_search.fit(X_train , y_train)
        best_model = grid_search.best_estimator_
    else : 
        best_model = model.fit(X_train , y_train)
    y_pred = best_model.predict(X_test)
    r2 = r2_score(y_test , y_pred)
    mse = mean_squared_error(y_test , y_pred)
    rmse = sqrt(mean_squared_error(y_test,y_pred))
    result.append({
        'model' : name , 
        'r2' : r2 , 
        'mse' : mse ,
        'rmse' : rmse ,
        'Model' : best_model 
    })
df_op = pd.DataFrame(result)
df_op
```

```python
import joblib
for res in result : 
    if res['model'] == 'SVM':
        print(res['Model'])
        joblib.dump(res['Model'] , 'best_model.pkl')
```
